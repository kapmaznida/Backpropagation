```python
# Verilen başlangıç ağırlıkları
W1 = 0.2
W2 = 0.3
W3 = 0.4
W4 = -0.1
W5 = 0.2
W6 = 0.1
W7 = 0.3
W8 = 0.2
W9 = 0.1

# Verilen input değerleri
X1 = 2
X2 = 1

# Gerçek çıktı
Y = 1

# Learning rate
lr = 0.5

#bias ihmal edilmiş olarak aldım. izlediğim derste de ödev verirken öyle diyordunuz sanırım.
# İleri yayılım işlemi
H1 = (X1 * W1) + (X2 * W4)
H2 = (X1 * W2) + (X2 * W5)
H3 = (X1 * W3) + (X2 * W6)
O1 = (H1 * W7) + (H2 * W8) + (H3 * W9)

# Hata hesaplama
delta_O1 = O1 - Y
delta_H1 = delta_O1 * W7
delta_H2 = delta_O1 * W8
delta_H3 = delta_O1 * W9

# Ağırlık güncelleme
W1 = W1 - lr * delta_H1 * X1
W2 = W2 - lr * delta_H2 * X1
W3 = W3 - lr * delta_H3 * X1
W4 = W4 - lr * delta_H1 * X2
W5 = W5 - lr * delta_H2 * X2
W6 = W6 - lr * delta_H3 * X2
W7 = W7 - lr * delta_O1 * H1
W8 = W8 - lr * delta_O1 * H2
W9 = W9 - lr * delta_O1 * H3

# Güncellenmiş ağırlıklar
print("Güncellenmiş ağırlıklar:")
print("W1: ", W1)
print("W2: ", W2)
print("W3: ", W3)
print("W4: ", W4)
print("W5: ", W5)
print("W6: ", W6)
print("W7: ", W7)
print("W8: ", W8)
print("W9: ", W9)
```
